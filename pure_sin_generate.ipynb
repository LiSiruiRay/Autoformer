```python
import math

def generate_sine_points(num_points):
    points = []
    for i in range(num_points):
        # x = x_max * (i / (num_points - 1))  # Scale x from 0 to x_max
        x = i
        y = math.sin(0.01 * i)
        points.append(y)
    return points

# Example usage:
num_points = 69680  # Number of points to generate
x_max = 2 * math.pi  # Maximum x-value (one full sine wave cycle)
points = generate_sine_points(num_points)
```

```python
import matplotlib.pyplot as pyplot
```

```python
pyplot.figure(figsize=(12, 6))
pyplot.plot(points)
pyplot.title('Time Series with pure sin')
pyplot.xlabel('Time')
pyplot.ylabel('Value')
pyplot.grid(True)
pyplot.show()
```

```python
def generate_stationary_data(generate_function, length: int, data_path:str):
    import pandas as pd
    result = generate_function(length)
    print(f"result: {result[:20]}")
    df = pd.read_csv('dataset/ETT-small/ETTm2.csv')
    cut_df = df[:length]
    cut_df['OT'] = result
    cut_df = cut_df[['date', 'OT']]
    cut_df.to_csv(data_path, index=False) #'dataset/self_made/sin_on_line_10k.csv'
    return cut_df
```
